# String in Python
A string is a sequence of characters enclosed inside quotes.
```python
name="Ramesh Das"
print(name)
```
Output:
Ramesh Das
```
**Strings are:**
1) Ordered
2) Immutable
3) Iterable
4) Unicode Supported

````python
# How to create String in Python
first_name="Ramesh"
last_name='Das'
profession='''Educator'''
print(first_name)
print(last_name)
print(profession)
````

````python
# String Concatenation
first_name="Ramesh"
last_name="Das"
profile_name=(first_name+" "+last_name)
print(profile_name)
````

````python
# String Repetition
name="Ankesh "
res=name*5
print(res)
````

````python
# Membership Operator in String
emailid="ramesh@gmail.com"
isCorrectEmailId="@gmail.com" in emailid # It will hold the boolean response of the condition
print(isCorrectEmailId)
````

````python
# String Indexing
name="Ramesh"
print(name[0])
print(name[-1])
print(name[-6])
#Forward Indexing starts from 0 and goes to n-1 if size of string is n
#Backward Indexing starts from -1 and goes to -n if size of string is n
````

````python
# String Slicing
name="Ramesh"
print(name[0:3]) # It like Range keyword in for loop. one is start idx and another is end idx.
print(name[2:]) # Here i write start idx but not have end idx thats why it print upto end.
print(name[:4]) # Here i write end idx but not have start idx thats why it print from start to end idx-1.
print(name[::-1]) # Here i write just reverse step so it print the string in reverse order like range in for loop.
````

````python
# Printing with the help of For Loop.
name="Ramesh"
for i in name:
    print(i)
````

````python
# String Length
text="Python is a great programming language, and it is widely used for web development, data analysis, artificial intelligence, and more."
text_length=len(text)
print(text_length)
````

````python
# String Conversion Methods
name="Ramesh Das"
small_letter_name=name.lower() # To convert all the characters of string into small letter.
capital_letter_name=name.upper() # To convert all the characters of string into capital letter.
proper_case_name=name.capitalize() # To convert the first character of string into capital letter.
title_case_name=name.title() # To convert the first character of each word into capital letter.
swapcase_name=name.swapcase() # To convert uppercase characters into lowercase and vice versa.
casefold_name=name.casefold() # Not recommended in Programming and it is similar it lower().
print(small_letter_name)
print(capital_letter_name)
print(proper_case_name)
print(title_case_name)
print(swapcase_name)
print(casefold_name)
````

````python
# Search and Find Methods in String
name="Ramesh Das"
print(name.find('R')) # It return the idx or the selected element.
print(name.find("Das")) # It also return the idx of full word as a signle element.
print(name.find('a')) # It takes left occurance of the selected element.
print(name.rfind('a')) # It takes right occurance of the selected element.
print(name.index('R')) # It like normal find but only diff that if not fount it return -1.
print(name.rindex('a')) # It like normal rfind but only diff that if not fount it return -1.
print(name.count('a')) # It return the count of the selected element in the string.
print(name.startswith('R')) # It return boolean response if the string start with the selected element.
print(name.endswith('s')) # It return boolean response if the string end with the selected element
````

````python
# Replace and Modify Methods in String
text1="I like Python programming language"
print(text1.replace("Python","Java")) # It replace the selected element with the new element.replace(what to replace, new element)
text2="   Hello, World!   "
print(text2.strip()) # It remove the leading and trailing white spaces from the string.
print(text2.lstrip()) # It remove the leading white spaces from the string.
print(text2.rstrip()) # It remove the trailing white spaces from the string.
r1=text1.removeprefix("I like ") # It remove the selected prefix from the string.
r2=r1.removesuffix(" programming language") # It remove the selected suffix from the string.
print(r2)
````

````python
# Split and Join Methods in String
text="Ramesh Das is an Educator"
split_text=text.split() # It split the string into list of words.
print(split_text)
split_from_right=text.rsplit() # It also split the string into list of words but it start splitting from right side.
print(split_from_right)
split_with_delimiter=text.split('a') # It split the string with the help of selected delimiter and it return list of splitted elements.
print(split_with_delimiter)
chunk_text=["Ramesh","Das","is","an","Educator"]
joined_text=' '.join(chunk_text) # It join the list of elements into string with the help of selected delimiter.
print(joined_text)
partioned_text=text.partition("Das") # It split the string into three parts first part is before the selected element, second part is the selected element and third part is after the selected element. It return a tuple of three parts.
print(partioned_text)
````

````python
# Checking Methods in String (Only return boolean response)
name="Ramesh Das"
print(name.isalnum()) # It return True if all the characters of string are alphanumeric.(alphabetic and numeric both)
print(name.isalpha()) # It return True if all the characters of string are alphabetic.(only alphabetic characters)
print(name.isdigit()) # It return True if all the characters of string are digit.(only numeric characters)
print(name.islower()) # It return True if all the characters of string are lowercase.
print(name.isupper()) # It return True if all the characters of string are uppercase.
print(name.isspace()) # It return True if all the characters of string are whitespace characters.
print(name.istitle()) # It return True if all the words of string start with capital letter and rest of the characters are small letter.
print(name.isidentifier()) # It return True if the string is a valid identifier.
print(name.isprintable()) # It return True if all the characters of string are printable characters.
````

````python
# String Alignment Methods
name="Ramesh"
print(name.center(20,"-"))
print(name.ljust(20,"-"))
print(name.rjust(20,"-"))
print(name.zfill(20))
````

````python
# String Formatting Methods
name="Ramesh"
age=22
print("My name is {} and my age is {}".format(name,age)) # It is a old method of string formatting and it is not recommended in programming.
print(f"My name is {name} and my age is {age}") # It is a new method of string formatting and it is recommended in programming. It is also called f-string formatting.
````

````python
# Translation and Mapping Methods in String
text="Ramesh Das is an Educator"
table=str.maketrans("aeiou","12345") # It create a translation table for the selected characters and their corresponding replacement characters.
translated_text=text.translate(table) # It translate the string with the help of translation table.
print(translated_text)
````

````python
# Some Escape Sequences in String
print("Code\nCoffee") # \n is used to print the string in new line.
print("Code\tCoffee") # \t is used to print the string with a tab space.
print("Code\\Coffee") # \\ is used to print the string with a backslash.
````
