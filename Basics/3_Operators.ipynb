# Operator (perform operations on variables) in Python:
1) Arithmetic Operators: +, -, *, /, //, %, **
2) Relational Operators: ==, !=, >, <, >=, <=
3) Logical Operators: and, or, not
4) Assignment Operators: =, +=, -=, *=, /=, //=, %=, **=
<!-- 5) Bitwise Operators: &, |, ^, ~, <<, >> -->
6) Membership Operators: in, not in
7) Identity Operators: is, is not

```python
# 1. Arithmatic Operators: +, -, *, /, //, %, **
a=10 #first variable
b=3 #second variable
print("a+b=",a+b) #Addition
print("a-b=",a-b) #Subtraction
print("a*b=",a*b) #Multiplication
print("a/b=",a/b) #Division
print("a//b=",a//b) #Floor Division
print("a%b=",a%b) #Modulus
print("a**b=",a**b) #Exponentiation
```

```python
#2. Relational Operators: ==, !=, >, <, >=, <= (Result is always boolean)
a=10
b=3
print("a==b:",a==b) #Equal to
print("a!=b:",a!=b) #Not Equal to
print("a>b:",a>b) #Greater than
print("a<b:",a<b) #Less than
print("a>=b:",a>=b) #Greater than or Equal to
print("a<=b:",a<=b) #Less than or Equal to
```

```python
# 3. Logical Operators: and, or, not ( here one cond   |  here another cond  )
left=True
right=False
print("left and right:",left and right) #Logical AND
print("left or right:",left or right) #Logical OR
print("not left:",not left) #Logical NOT
```

# Truth Table of AND, OR, NOT:
A | B | A AND B | A OR B | NOT A | NOT B
--|---|---------|--------|------|------
True | True | True | True | False | False
True | False | False | True | False | True
False | True | False | True | True | False
False | False | False | False | True | True

```python
#4. Assignment Operators: =, +=, -=, *=, /=, //=, %=, **=
x=5
print("Initial value of x:",x)
x+=3 # Shorthand Operators (x = x + 3)
print("After x+=3, x:",x)
```

```python
# All Shorthand Operators:
x=5
print("Initial value of x:",x)
x+=3 # x = x + 3
print("After x+=3, x:",x)
x-=2 # x = x - 2
print("After x-=2, x:",x)
x*=4 # x = x * 4
print("After x*=4, x:",x)
x/=2 # x = x / 2
print("After x/=2, x:",x)
x//=3 # x = x // 3
print("After x//=3, x:",x)
x%=2 # x = x % 2
print("After x%=2, x:",x)
x**=3 # x = x ** 3
print("After x**=3, x:",x)
```

```python
#6. Membership Operators: in, not in
fruits=["apple","banana","cherry"]
print("apple" in fruits) # True
print("mango" in fruits) # False
print("mango" not in fruits) # True
```

```python
#7. Identity Operators: is, is not
a=10
b=10
print("a is b:",a is b) # True (both refer to the same object in memory)
print("a is not b:",a is not b) # False
```
