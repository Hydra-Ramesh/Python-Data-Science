# Variable in Python
A variable in Python is a named location in memory that is used to store data. It acts as a container for values that can be accessed and manipulated throughout the program. Variables are created by assigning a value to a name using the assignment operator (=). The value assigned to a variable can be of various data types, such as integers, floats, strings, lists, dictionaries, etc. Once a variable is defined, you can use it to perform operations, store results, and manage data within your Python programs.

```python
age=22
height=5.9
name="Ramesh Das"
is_student=True

print(age)
print(height)
print(name)
print(is_student)
```

```python
# Python is Case Sensitive that means A is not same as a.
a=10
A=20
print(a)
print(A)
print(a is A)
```

# L Value in Python
In Python, an L-value (Left-value) refers to a variable or an expression that can appear on the left side of an assignment statement. It is a location in memory that can be assigned a value. For example, in the statement `x = 10`, `x` is an L-value because it can be assigned the value `10`. L-values are typically variables, but they can also include attributes of objects or elements of data structures that can be assigned values.
# R Value in Python
In Python, an R-value (Right-value) refers to a value or an expression that can appear on the right side of an assignment statement. It is a value that can be assigned to an L-value. For example, in the statement `x = 10`, `10` is an R-value because it is the value being assigned to the variable `x`. R-values can be literals (like numbers or strings), variables, or more complex expressions that evaluate to a value.

# Type Checking in Python
In Python, you can check the type of a variable using the built-in `type()` function. This function takes a variable as an argument and returns its data type.


```python
name="Ramesh Das"
print(name)
type(name)
```

```python
age=22
height=5.9
name="Ramesh Das"
is_student=True

print(age)
print(height)
print(name)
print(is_student)
print(type(age))
print(type(height))
print(type(name))
print(type(is_student))
```

# Datatypes in Python
Defination:
Data types are a classification of data that tells the compiler or interpreter how the programmer intends to use the data.
Importance of Data types:
1) Data type ensure that data is stored in an efficient way in memory.
2) Data type help in performing operations on data. For example, you can perform mathematical operations on integers and floats, but not on strings.
3) Data type help in preventing errors. For example, if you try to perform a mathematical operation on a string, you will get an error.

```python
#1. Number: int, float, complex
num1=10 #int
num2=3.14 #float
num3=2+3j #complex
print(num1)
print(num2)
print(num3)
print(type(num1))
print(type(num2))
print(type(num3))
```

```python
#2. Boolean: bool
isPaid=True
isActive=False
print(isPaid)
print(isActive)
print(type(isPaid))
print(type(isActive))
```

```python
a=20
b=20
c=a==b
print(c)
print(type(c))
```

```python
# 3. String: str
name="Ramesh Das"
print(name)
print(type(name))
print(name+"(Babu)") #Concatenation
```

```python
# 4. List: list (Mutable)
number=[1,2,3,4,5]
print(number)
print(type(number))
```

```python
# 5. Tuple: tuple (Immutable)
numbers=(1,2,3,4,5)
print(numbers)
print(type(numbers))
```

```python
# 6. Dictionary: dict (Mutable)
user={
    "name":"Ramesh Das",
    "age":22,
    "is_student":True
}
print(user)
print(type(user))
```

```python
# 7. Set: set (Mutable)
unique_numbers={1,2,3,4,4,5}
print(unique_numbers)
print(type(unique_numbers))
```

```python
# 8. NoneType: None
value=None
print(value)
print(type(value))
```
