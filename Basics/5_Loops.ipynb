# Loops in Python
1) For Loop
2) While Loop
3) Nested Loops
4) Loop Control Statements (break, continue, pass)

```python
range(5,20) # range(start, stop, step)
```

```python
for i in range(10):
    print(i)
```

```python
for i in range(5,20):
    print(i)
```

```python
for i in range(5,20,2):
    print(i)
```

```python
for i in range(20,5,-2):
    print(i)
```

```python
str="Ramesh Das"
for i in str:
    print(i)
```

```python
# While Loop
# init
# while (condition){
#     body
#     update
# }
i=0
while i<10:
    print(i)
    i+=2
```

```python
# Break Condition
for i in range(10):
    print(i)
    if i==5:
        break
```

```python
# Continue Keyword
for i in range(20):
    if i%3==0 or i%5==0:
        continue
    print(i)
```

```python
# Pass Keyword (it does nothing, just a placeholder) it perform null operation
for i in range(10):
    pass
    print(i)
```

```python
# Nested Loops
for i in range(5):
    for j in range(5):
        print(f"{i},{j}",end=" ")
    print()

# i=0 => j=0,1,2,3,4
# i=1 => j=0,1,2,3,4
# i=2 => j=0,1,2,3,4
# i=3 => j=0,1,2,3,4
# i=4 => j=0,1,2,3,4
```

```python
# WAP to find the sum of first n natural numbers
n=int(input("Enter n: "))
sum=0
i=n
while i>0:
    sum+=i;
    i-=1
print("Sum:",sum)
```

```python
# Pattern Printing
# *
# **
# ***
# ****
# *****
n=int(input("Enter n: "))
for i in range(n):
    for j in range(i+1):
        print("*",end="")
    print() # It will print a new line after each row
```
