## What is Python?
Python is a High Level Language, Interpreted Programming Language known for Simple Syntax and Readability.
## What is Syntax?
Syntax refers to the set of rules that define the combinations of symbols that are considered to be correctly structured in a programming language. It dictates how code should be written and organized to be understood by the interpreter or compiler. In Python, for example, indentation is a crucial part of the syntax, as it defines the scope of loops, functions, and other code blocks.
## What is Semantics?
Semantics refers to the meaning of the code, or what the code does when it is executed. It is concerned with the logic and behavior of the program, rather than just its structure. In Python, for example, the semantics of a loop would involve how it iterates over a sequence and what actions it performs during each iteration.

# Indentation in Python
Indentation in Python is used to define the scope of loops, functions, and other code blocks. It is a crucial part of the syntax and is typically done using four spaces or a single tab. Proper indentation is essential for the correct execution of Python code, as it indicates which statements belong to which blocks of code. For example, in a loop or a function, all the statements that are part of that block must be indented at the same level. Incorrect indentation can lead to syntax errors or unintended behavior in the program.

```python
age=17
if age>18:
    print(age)

print(age)
```

**How Many Characters are there in Python?**
1) Letters: (a-z, A-Z)
2) Digits: (0-9)
3) Special Characters: (e.g., !, @, #, $, %, ^, &, *, (, ), -, +, =, {, }, [, ], |, \, :, ;, ", ', <, >, ?, /)

**Tokens In Python**
1) Keywords: These are reserved words that have a specific meaning in Python and cannot be used as identifiers (e.g., if, else, for, while, def, class).
2) Identifiers: These are names used to identify variables, functions, classes, and other entities in Python. They must follow certain rules (e.g., they cannot start with a digit, cannot contain spaces, and cannot be the same as keywords).
3) Literals: These are fixed values that are directly represented in the code (e.g., numbers, strings, booleans).
4) Operators: These are symbols that perform operations on variables and values (e.g., +, -, *, /, =).
5) Delimiters or Punctuation: These are symbols that are used to define the structure of the code (e.g., parentheses, brackets, braces, commas, colons).

**Naming Convention in Any Programming Language with Best Practices**
1) Kabab Case: Every word is Capitalized. Example: CONSTANT_VALUE Used for Constants.
2) Snake Case: Every word is in lowercase and separated by an underscore. Example: user_name Used for Variables and Functions.
3) Camel Case: The first word is in lowercase and the first letter of each subsequent word is capitalized. Example: constantValue Used for Variables and Functions.
4) Pascal Case: Every word is capitalized. Example: ConstantValue Used for Classes.

Best Practices:
1) Use meaningful and descriptive names for variables, functions, and classes to enhance code readability.
2) Avoid using single-letter variable names except for loop counters.
3) Use consistent naming conventions throughout the codebase.
4) Do not use reserved keywords as identifiers.

# Literals in Python
Literals in Python are fixed values that are directly represented in the code. They can be of various types, including:
1) Numeric Literals: These represent numbers and can be of different types such as integers (e.g., 42), floating-point numbers (e.g., 3.14), and complex numbers (e.g., 1 + 2j).
2) String Literals: These represent sequences of characters and are enclosed in single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """). Example: 'Hello, World!'
3) Boolean Literals: These represent the truth values and can be either True or False. Example: is_valid = True
4) None Literal: This represents the absence of a value and is denoted by None. Example: result = None
Literals are used to assign values to variables and to represent data in the code. They are an essential part of programming as they allow developers to work with fixed values and data in their programs.

```python
a=3
b=3.14
c=2+3j
d=True
e=None
f='Hello, World!'
```
