```python
# IF Statement
# You are building one ECI registration form, and user will give their age just tell them they are eligible for registration or not. If they are eligible then print "You are eligible for registration" otherwise print "You are not eligible for registration". (Eligibility criteria is age should be greater than or equal to 18 years)
age=int(input("Enter your age: "))
if age>=18:
    print("You are eligible for registration")
if age<18:
    print("You are not eligible for registration")
```

```python
# Grade Calculator of Korea
marks=int(input("Enter your Marks:"))
if marks>90:
    print("Excellent")
if marks>80:
    print("Very Good")
if marks>70:
    print("Good")
else:
    print("Bad")
```

```python
# If-Elif-Else
# Grade Calculator of Korea
marks=int(input("Enter your Marks:"))
if marks>90:
    print("Excellent")
elif marks>80:
    print("Very Good")
elif marks>70:
    print("Good")
else:
    print("Bad")
```

```python
# Name according to the age
age=int(input("Enter your age:"))
if age<2:
    print("Baby")
elif age<12:
    print("Child")
elif age<19:
    print("Teenager")
elif age<60:
    print("Adult")
else:
    print("Senior Citizen")
```

```python
# WAP to check a number is positive or negative as well as even or odd.
num=int(input("Enter a number:"))
if num>=0:
    print("Positive")
    if num%2==0:
        print("Even")
    else:
        print("Odd")
else:
    print("Negative")
    if num%2==0:
        print("Even")
    else:
        print("Odd")
```
